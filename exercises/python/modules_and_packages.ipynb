## os
Write a Python program to print the list of files in the current folder and their absolute path.

```python
import os
current_folder = os.getcwd()
items=os.listdir(current_folder)
for item in items:
  abs_path=os.path.abspath(item)
  print(f"{item}: {abs_path}")
```

## math
Write a Python program that finds the maximum of log$_2 10000$ and $ e^e $.

```python
import math
log2_10000=math.log2(10000)
e_to_e=math.exp(math.e)
max_value=max(log2_10000,e_to_e)
print(f"log2(10000)={log2_10000}")
print(f
      "e^e={e_to_e}")

print(f"Max value: {max_value}")
```
