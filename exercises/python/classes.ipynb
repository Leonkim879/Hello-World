## Course Class
Define a course class with the followings:
- A `code` attribute.
- A `name` attribute.
- A `credits` attribute.
- A `semester` attribute.
- An `instructor` attribute.
- An `__init__` method.
- A `__str__` method.

```python
class Course:
  def __init__(self,code,name,credits,semester="",instructor=""):
    self._code = code
    self._name = name
    self._credits = credits
    self._semester = semester
    self._instructor = instructor
  def __str__(self):
    s = "==========Course=======\n"
    s += f"name: {self._code}\n"
    s += f"code: {self._name}\n"
    s += f"credits: {self._credits}\n"
    return s

```

```python
python101 = Course("BP101","Python101","WS105","Mehran")
print(python101)
```
